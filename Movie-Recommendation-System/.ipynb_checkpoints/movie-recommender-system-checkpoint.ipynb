```python
import numpy as np
import pandas as pd
```

```python
movies_df = pd.read_csv("tmdb_5000_movies.csv")
credits_df = pd.read_csv("tmdb_5000_credits.csv")
```

```python
#MERGING 2 DATAFRAMES

movies_df = movies_df.merge(credits_df, on = "title")
```

```python
movies_df.shape
```

```python
movies_df.info()
```

```python
#genre
#id
#original language
#keywords
#title
#overview
#release date(maybe later)
#cast
#crew

movies_df = movies_df[["genres", "movie_id", "keywords", "original_language", "title", "overview", "cast", "crew"]]
```

```python
movies_df.info()
```

```python
movies_df.isnull().sum()
```

```python
movies_df.dropna(inplace = True)
```

```python
movies_df.iloc[0].genres
```

```python
import ast
def process_dict_to_list(obj):
    obj = ast.literal_eval(obj) #string ---> dict
    L = []
    for i in obj:
        L.append(i["name"])
    return L
```

```python
process_dict_to_list(movies_df.iloc[0].genres)
```

```python
movies_df["genres"] = movies_df["genres"].apply(process_dict_to_list)
```

```python
movies_df["keywords"] = movies_df["keywords"].apply(process_dict_to_list)
```

```python
movies_df
```

```python
def top_actors(movie_cast):
    movie_cast = ast.literal_eval(movie_cast)
    actors = []
    for char_desc in movie_cast:
        actors.append(char_desc["name"])
        if len(actors) == 3:
            return actors
    return actors     
```

```python
movies_df["cast"] = movies_df["cast"].apply(top_actors)
```

```python
def only_director(movie_crew):
    movie_crew = ast.literal_eval(movie_crew)
    for crew_desc in movie_crew:
        if crew_desc["job"] == "Director":
            return [crew_desc["name"]]
```

```python
only_director(movies_df["crew"][0])
```

```python
movies_df["crew"] = movies_df["crew"].apply(only_director)
```

```python
movies_df
```

```python
movies_df.isnull().sum()
```

```python
null_only = movies_df[movies_df.isnull().any(axis = 1)]
null_only
```

```python
movies_df["overview"] = movies_df["overview"].apply(lambda x: x.split())
```

```python
movies_df.dropna(inplace = True)
```

```python
movies_df["genres"] = movies_df["genres"].apply(lambda l : [i.replace(" ", "") for i in l])
movies_df["keywords"] = movies_df["keywords"].apply(lambda l : [i.replace(" ", "") for i in l])
movies_df["cast"] = movies_df["cast"].apply(lambda l : [i.replace(" ", "") for i in l])
movies_df["crew"] = movies_df["crew"].apply(lambda l : [i.replace(" ", "") for i in l])
```

```python
movies_df.shape
```

```python
movies_df["tags"] = movies_df["overview"] + movies_df["genres"] + movies_df["keywords"] + movies_df["crew"]
```

```python
movies_df.head(1)
```

```python
new_df = movies_df[["movie_id", "title", "tags"]]
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x: " ".join(x))
```

```python
#converting to lower case
new_df["tags"] = new_df["tags"].apply(lambda x: x.lower())
```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = "english")
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors 
```

```python
import nltk
```

```python
#for stemming words eg: [danced, dancing, dance] = [dance, dance, dance]
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem_similar(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(stem_similar)
```

```python
new_df.head(3)
```

```python
new_df.shape
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
new_df[new_df["title"] == "Batman Begins"].index[0]
```

```python
def recommend(movie_title):
    movie_index = new_df[new_df["title"] == movie_title].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key=lambda x: x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    

```

```python
recommend("The Truman Show")
```
